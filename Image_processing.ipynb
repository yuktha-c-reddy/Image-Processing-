#Import the necessary lib , open cv and matplotlib mainly

```python
import pandas as pd
import numpy as np
from glob import glob
import cv2
import matplotlib.pylab as plt
import os
```

#Read the images

```python
cats=glob('/content/imageprocessing/cats/*.jpg')
dogs=glob('/content/imageprocessing/dogs/*.jpg')
```

#using plt and cv2 - gives a np array for image

```python
img_mpl =plt.imread(cats[0])
img_cv =cv2.imread(cats[0])
```

```python
pd.Series(img_mpl.flatten()).plot(kind="hist",bins=50,title="Pixels")
plt.show()
```

#Display Image

```python
fig , ax = plt.subplots(figsize=(10,10))
ax.imshow(img_mpl)
ax.axis('off')
plt.show()
```

#RGB representation

```python
fig , axs =plt.subplots(1,3,figsize=(10,10))
axs[0].imshow(img_mpl[:,:,0],cmap="Reds")
axs[1].imshow(img_mpl[:,:,1],cmap="Greens")
axs[2].imshow(img_mpl[:,:,2],cmap="Blues")
axs[0].set_title("RED")
axs[1].set_title("GREEN")
axs[2].set_title("BLUE")
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
plt.show()
```

#MATPLOTLIB - RGB   , CV - BGR

```python
fig , axs = plt.subplots(1,2,figsize=(10,10))
axs[0].imshow(img_mpl)
axs[0].axis('off')
axs[0].set_title('Matplotlib Image')
axs[1].imshow(img_cv)
axs[1].axis('off')
axs[1].set_title('CV Image')
plt.show()
```

#BGR -> RGB

```python
img_cvrgb = cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
fig , ax = plt.subplots(figsize=(7,7))
ax.imshow(img_cvrgb)
ax.axis('off')
ax.set_title("Converted")
plt.show()
```
